# Toplulaştırma ve Gruplama
- Toplulaştırma -> Özet istatistikler(mean, max, std..)
count() -> eleman sayısı<br>
first() -> ilk eleman<br>
last() -> son eleman <br>
mean() -> ortalama<br>
median() -> ortanca<br>
min() -> minimum<br>
max() -> maksimum<br>
std() -> standart sapma<br>
var() -> varyans<br>
sum() -> toplam<br> 


```python
df.head()
```

```python
df.groupby("sex").agg({"age": "mean"}) # Cinsiyete göre yaş ortalaması
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"]})
```

```python
df.groupby(["sex", "embark_town"]).agg({"age": ["mean", "sum"],
                                        "survived": "mean"})
```

```python
df.groupby(["sex", 
            "embark_town", "class"]).agg({"age": ["mean", "sum"],
                                          "survived": ["mean", "count"]})
```

# Pivot Table
- Groupby işlemlerine benzer olarak veri setine kırılmlar açısından değerlendirmek ve ilgilenen özet istatistiği değerlerini bu kırılımlarda değerlendirme imkanı sağlar.

```python
df.pivot_table("survived", "sex", "embarked") # ortalama, satır, sütun
```

```python
df.pivot_table("survived", "sex", "embarked", aggfunc="std") # standart sapma
```

```python
df.pivot_table("survived", "sex", ["embarked", "class"]) 
```

# Apply & Lambda
- Apply döngülerden daha pratik satır veya sütunlarda otomatik olarak çalışır.
- Lambda kullan at fonksiyondur.

```python
df.head()
```

```python
# Yeni değişkenler oluşturalım
df["age2"] = df["age"] * 2
df["age3"] = df["age"] * 5
df.head()
```

```python
# Amaç yaş değişkenlerinin 10'a bölünmesi
# df["age"] / 10 
# df["age2"] / 10
# don't repeat yourself ! 

# 1. Yol -> Klasik Yöntem Döngü ile
for col in df.columns:
  if "age" in col:
    df[col] = df[col] / 10  

df.head()
```

```python
df[["age", "age2", "age3"]].apply(lambda x: x ** 2)
```

```python
# Değişken isimlerini programatik olarak seçmek
df.loc[:, df.columns.str.contains("age")].apply(lambda x: x / 10).head()
```

```python

df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x - x.mean()) / x.std()).head()
```

```python
def standart_scaler(col_name):
  return (col_name - col_name.mean()) / col_name.std()
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

- Apply fonksiyonu bize satırlarda veya sütunlarda elimizdeki bir fonksiyonu uygulama imkanı sağlar.

```python
# İşlemi kaydetmek için
df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)
df.head()
```

# Birlşetirme İşlemleri
1. Concat
2. Merge

```python
import numpy as np

m = np.random.randint(1, 30, size=(5,3))
df1 = pd.DataFrame(m, columns=["var1", "var2", "var3"])
df2 = df1 + 99
```

- 2 veri setini alt alta birleştirmek için,

```python
pd.concat([df1, df2]) #liste içinde gönderilir
# İndexlerde problem var
```

```python
pd.concat([df1, df2], ignore_index=True)
```

- Merge biraz daha detaylı birleştirme imkanı sunar

```python
df1 = pd.DataFrame({"employees": ["john", "dennis", "mark", "maria"],
                    "group": ["accounting", "engineering", "engineering", "hr"]})
df2 = pd.DataFrame({"employees": ["john", "dennis", "mark", "maria"],
                    "start_date": [2010, 2009, 2014, 2019]})
pd.merge(df1, df2)
# Otomatik olarak birleştirdi
```

```python
df3 = pd.merge(df1, df2, on="employees")
df4 = pd.DataFrame({"group": ["accounting", "engineering", "hr"],
                    "manager": ["Caner", "Mustafa", "Berkcan"]})
pd.merge(df3, df4, on="group")
```
