# Crm Analitiğine Giriş (Müşteri İlişkileri Yönetimi)
- Bir şirketin müşterileri ile olan ilişkilerinin yönetimi söz konusudur. Müşteriler ile ne tür ilişkiler kurulur?
- CRM müşteriler ile kurulucak olan iletişimi farklı açılardan, yüksek farkındalıklar ile gerçekleştirme imkanı sağlar.
- KPI (Temel performans Göstergeleri)
- Bazı CRM Kavramları:
1. Customer Lifecycle (Müşteri yaşam döngüsü optimizasyonları)
    - Müşterilerin şirketle olan süreçlerini hem görsel hem de takip etme anlamında tasarlayan süreçlerdir. (Müşteri siteye girer, üye olur, ürün satın alır bunların hepsi bir süreçtir.)
2. İletişim (Dil, Renk, Görseller, Kampanyalar)
3. Müşteri Bulma Çalışmaları
4. Müşteriyi Elde Etme Çalışmaları (Churn)
5. Çapraz Satış, Üst Satış(Küçük boy yerine büyük boy)
6. Müşteri Segmentasyonu 
    - Müşterilerin hepsini yanı kabul etmek yerine segmentlere ayırmak
<br>
#### Crm analitiği çalışmaları ile amacımız tüm müşteri ilişkileri sürecini veriye dayalı olarak daha verimli hale getimeye çalışmaktır.


# KPI (Key Performance Indicators - Temel Performans Göstergeleri)
- Şirket, departman ya da çalışanların performanslarını değerlendirmek için kullanılan matematiksel göstergelerdir.
- Örnekler:
1. Customer Acquisition Rate (Müşteri Kazanma Oranı)
    - Belirli bir zamanda kazanılan müşteri yüzdesidir.
2. Customer Retention Rate (Müşteri Elde Tutma Oranı)
    - Müşteri 2. alışverişinden sonra artık firmanın müşterisi olarak düşünülebilir. 2. alışverişten sonra belirli bir zaman periyodunda ne kadar müşteri alışveriş yapmayı sürdürdü? Retention Rate
3. Customer Churn Rate (Müşteri Terk Oranı)
4. Conversion Rate (Dönüşüm Oranı)
    - Bir ilanı gören ile satın alan arasındaki oran
5. Growth Rate (Büyüme Oranı) <br><br/>
