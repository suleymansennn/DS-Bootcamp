```python
l = [2, 13, 18, 93, 22]

def odd_even(list):
    odd_list = []
    even_list = []
    for i in list:
        if i % 2 == 0:
            even_list.append(i)
        else:
            odd_list.append(i)
    return even_list, odd_list

even_list, odd_list = odd_even(l)
print(even_list, odd_list)
```
